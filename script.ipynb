# [dbo].[GetAllIndicators]

```sql
/****** Object:  StoredProcedure [dbo].[GetAllIndicators]    Script Date: 6/5/2021 7:01:05 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[GetAllIndicators]
	 @UniversityId int
AS
    BEGIN
    select
    Ind.Indicator_Id,
    Ind.University_Id,
    IndR.IndicatorName,
    Ind.[Value],
    Ind.UnitOfMeasure,
    Univ.UniversityName,
    Ind.Year
    from Indicators as Ind
    join Universities as Univ on Ind.University_Id = Univ.University_Id
    join Indicator as IndR on Ind.Indicator_Id = IndR.Indicator_Id
    where Ind.University_Id = @UniversityId
END

GO
```
